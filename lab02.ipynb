```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.svm import SVC
from scikitplot import metrics# import from sciketplot instead of sklearn
from sklearn.linear_model import SGDClassifier

FilePath = r'DryBeanDataset\Dry_Bean_Dataset.xlsx'
```

```python
DryBeanData = pd.read_excel(FilePath)# seems that my panda version works.
```

```python
#slicing
DBData = DryBeanData.loc[:,['Area','Perimeter','MajorAxisLength','MinorAxisLength','ConvexArea','Solidity','Class']]
```

```python
pd.plotting.scatter_matrix(DBData,alpha=1,figsize=(8,10))
plt.show()
```

```python
#2
print(DBData.Class.describe())
print("-----------------------------------")
print(DBData.Class.unique())
print("-----------------------------------")
print(DBData.Class.value_counts())
```

```python
classBias = DBData.Class.value_counts()
```

```python
classBias.sort_values().plot.pie()
```

```python
#3
x_train,x_test,y_train,y_test = train_test_split(
    DBData.iloc[:,0:6],DBData.iloc[:,6:],train_size = 0.8,
    random_state = 0) 
```

```python
x_train.info()#10888 instances
```

```python
#feature 
x_train.head(5)#from 0.9 to 70000, not on a similar scale
```

```python
# make sure test and train set are transformed by the same scaler
sScaler = StandardScaler().fit(DBData.iloc[:,0:6]) # by using full data set
x_train_S = sScaler.transform(x_train)
x_test_S = sScaler.transform(x_test)
```

```python
#5 SVM
svmModel = SVC(C = 10, kernel = 'rbf',gamma = 'auto',decision_function_shape='ovo')
# fit training data
svmClf = svmModel.fit(x_train_S,y_train.values.ravel())
```

```python
scoreSVM = svmClf.score(x_test_S,y_test)
print("The score of SVM is : %f"%scoreSVM)
```

```python
#make prediction for confusion matrix
svmPred = svmClf.predict(x_test_S)
beanLabels = DBData.Class.unique().tolist()
# confusion matrix
#svmConMatrix = metrics.confusion_matrix(y_test,svmPred,labels = beanLabels)
#print(svmConMatrix)
```

```python
# plot_confusion_matrix from scikitplot has more parameters than it from sklearn
metrics.plot_confusion_matrix(
    y_test,svmPred,normalize='pred',
    cmap = 'Blues',
    figsize = (9,10),
    hide_zeros = 'true')
```

```python
# 6 SGD classifier 
sgdClf = SGDClassifier(loss = 'log',penalty='elasticnet')
sgdClf.fit(x_train_S,y_train.values.ravel())
#score
scoreSGD = sgdClf.score(x_test_S,y_test)
print("The score of SGD is : %f"%scoreSGD)
#make prediction for confusion matrix
sgdPred = sgdClf.predict(x_test_S)
metrics.plot_confusion_matrix(
    y_test,sgdPred,normalize='pred',
    cmap = 'Blues',
    figsize = (9,10),
    hide_zeros = 'true')
```
